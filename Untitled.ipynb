```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from datetime import datetime

# Parameters for South Africa
WIDTH, HEIGHT = 512, 512

# South Africa's bounding box coordinates (normalized to texture space)
SA_BOUNDS = {
    "min_lon": 16.45, "max_lon": 32.89,  # Longitude range
    "min_lat": -34.83, "max_lat": -22.13  # Latitude range
}

# Major South African cities and weather stations
SA_WEATHER_STATIONS = [
    {"name": "Johannesburg", "x": 0.55, "y": 0.4, "elevation": 1753, "temp_factor": 0.8},
    {"name": "Pretoria", "x": 0.57, "y": 0.42, "elevation": 1339, "temp_factor": 0.85},
    {"name": "Durban", "x": 0.7, "y": 0.25, "elevation": 5, "temp_factor": 1.1},
    {"name": "Cape Town", "x": 0.3, "y": 0.15, "elevation": 5, "temp_factor": 0.9},
    {"name": "Bloemfontein", "x": 0.5, "y": 0.5, "elevation": 1395, "temp_factor": 0.75},
    {"name": "Port Elizabeth", "x": 0.6, "y": 0.1, "elevation": 15, "temp_factor": 1.0},
    {"name": "Nelspruit", "x": 0.65, "y": 0.35, "elevation": 619, "temp_factor": 1.2}
]

def generate_sa_temperature_texture(month):
    """Generate temperature texture based on month (0-11)"""
    x, y = np.meshgrid(np.linspace(0, WIDTH-1, WIDTH), np.linspace(0, HEIGHT-1, HEIGHT))
    
    # Base temperature pattern
    temp = np.zeros((HEIGHT, WIDTH))
    
    # Seasonal variation (summer = Dec-Feb, winter = Jun-Aug)
    is_summer = month in [11, 0, 1]
    base_temp = 25 if is_summer else 15  # Average base temperature
    
    # Latitude effect (colder south)
    lat_effect = (1 - y/HEIGHT) * 10
    
    # Elevation effect (colder higher)
    elevation_map = generate_elevation_map()
    elevation_effect = elevation_map * 0.1  # -1째C per 100m
    
    # Add temperature sources for each city
    for station in SA_WEATHER_STATIONS:
        tex_x = station["x"] * WIDTH
        tex_y = station["y"] * HEIGHT
        dist = np.sqrt((x - tex_x)**2 + (y - tex_y)**2)
        
        # Station temperature influenced by season and local factor
        station_temp = base_temp * station["temp_factor"]
        temp += station_temp * np.exp(-dist / (50 + np.random.random() * 30))
    
    # Combine effects
    temp = temp / len(SA_WEATHER_STATIONS)  # Average station influences
    temp -= lat_effect  # Colder in the south
    temp -= elevation_effect  # Colder at higher elevations
    
    # Add noise for realism
    noise = np.random.normal(0, 0.5, (HEIGHT, WIDTH))
    temp = np.clip(temp + noise, 0, 40)  # Clip to reasonable temps
    
    # Normalize to 0-255
    temp_normalized = ((temp - 0) / (40 - 0) * 255).astype(np.uint8)
    img = Image.fromarray(temp_normalized)
    img.save(f"sa_temperature_{month+1}.png")
    print(f" Saved South Africa temperature texture for month {month+1}")
    return img

def generate_sa_precipitation_texture(month):
    """Generate precipitation texture based on month (0-11)"""
    x, y = np.meshgrid(np.linspace(0, WIDTH-1, WIDTH), np.linspace(0, HEIGHT-1, HEIGHT))
    
    precip = np.zeros((HEIGHT, WIDTH))
    
    # Seasonal patterns (summer rainfall in north, winter rainfall in southwest)
    is_summer = month in [11, 0, 1]
    is_winter = month in [5, 6, 7]
    
    # Base precipitation patterns
    if is_summer:
        # Summer rainfall in northeast
        precip += 50 * np.exp(-((x - WIDTH*0.7)**2 + (y - HEIGHT*0.3)**2) / (100**2))
        precip += 30 * np.exp(-((x - WIDTH*0.6)**2 + (y - HEIGHT*0.4)**2) / (120**2))
    elif is_winter:
        # Winter rainfall in southwest (Cape)
        precip += 60 * np.exp(-((x - WIDTH*0.3)**2 + (y - HEIGHT*0.15)**2) / (80**2))
    else:
        # Transition seasons - lighter rain
        precip += 20 * np.exp(-((x - WIDTH*0.5)**2 + (y - HEIGHT*0.3)**2) / (150**2))
    
    # Orographic precipitation (windward slopes)
    elevation_map = generate_elevation_map()
    precip += elevation_map * 0.5  # More rain on slopes
    
    # Add noise for realism
    noise = np.random.normal(0, 5, (HEIGHT, WIDTH))
    precip = np.clip(precip + noise, 0, 100)  # Clip to 0-100mm
    
    # Normalize to 0-255
    precip_normalized = (precip / 100 * 255).astype(np.uint8)
    img = Image.fromarray(precip_normalized)
    img.save(f"sa_precipitation_{month+1}.png")
    print(f" Saved South Africa precipitation texture for month {month+1}")
    return img

def generate_sa_wind_texture(month):
    """Generate wind texture based on month (0-11)"""
    x, y = np.meshgrid(np.linspace(0, WIDTH-1, WIDTH), np.linspace(0, HEIGHT-1, HEIGHT))
    
    # Seasonal wind patterns
    is_summer = month in [11, 0, 1]  # December-February
    
    if is_summer:
        # Summer: Southeast trade winds
        u = np.ones((HEIGHT, WIDTH)) * 0.8   # East component
        v = np.ones((HEIGHT, WIDTH)) * -0.6  # South component
    else:
        # Winter: Northwest winds
        u = np.ones((HEIGHT, WIDTH)) * -0.7  # West component
        v = np.ones((HEIGHT, WIDTH)) * 0.5  # North component
    
    # Topographic effects
    elevation_map = generate_elevation_map()
    terrain_effect = np.gradient(elevation_map)
    u += terrain_effect[1] * 0.2
    v += terrain_effect[0] * 0.2
    
    # Coastal effects
    coastal_mask = np.where(
        (x < WIDTH*0.1) | (x > WIDTH*0.9) | (y < HEIGHT*0.1) | (y > HEIGHT*0.9),
        1.0, 0.3
    )
    coastal_turbulence = (
        np.sin(2*np.pi*x/WIDTH*4) * 
        np.cos(2*np.pi*y/HEIGHT*3) * 
        coastal_mask
    )
    u += coastal_turbulence * 0.3
    v += coastal_turbulence * 0.2
    
    # Normalize to [0,255]
    r = ((u + 1) / 2 * 255).astype(np.uint8)  # East-West
    g = ((v + 1) / 2 * 255).astype(np.uint8)  # North-South
    b = np.zeros_like(r)                      # Unused
    
    wind_img = np.stack([r, g, b], axis=2)
    img = Image.fromarray(wind_img, 'RGB')
    img.save(f"sa_wind_{month+1}.png")
    print(f" Saved South Africa wind texture for month {month+1}")
    return img

def generate_elevation_map():
    """Generate a simplified elevation map for South Africa"""
    x, y = np.meshgrid(np.linspace(0, WIDTH-1, WIDTH), np.linspace(0, HEIGHT-1, HEIGHT))
    
    # Highveld region (eastern plateau)
    elevation = 1000 * np.exp(-((x - WIDTH*0.6)**2 + (y - HEIGHT*0.4)**2) / (150**2))
    
    # Drakensberg mountains
    elevation += 2000 * np.exp(-((x - WIDTH*0.65)**2 + (y - HEIGHT*0.3)**2) / (50**2))
    
    # Cape Fold Mountains
    elevation += 1500 * np.exp(-((x - WIDTH*0.3)**2 + (y - HEIGHT*0.15)**2) / (40**2))
    
    # Coastal areas are low
    elevation -= 500 * np.exp(-((x - WIDTH*0.1)**2 + (y - HEIGHT*0.1)**2) / (200**2))
    elevation -= 500 * np.exp(-((x - WIDTH*0.9)**2 + (y - HEIGHT*0.9)**2) / (200**2))
    
    # Clip to reasonable values
    elevation = np.clip(elevation, 0, 3000)
    
    return elevation / 3000  # Normalize to 0-1

def generate_sa_pressure_texture(month):
    """Generate atmospheric pressure texture"""
    x, y = np.meshgrid(np.linspace(0, WIDTH-1, WIDTH), np.linspace(0, HEIGHT-1, HEIGHT))
    
    # Seasonal pressure patterns
    is_summer = month in [11, 0, 1]
    
    if is_summer:
        # Summer: Low pressure over interior
        pressure = 1010 + 10 * np.exp(-((x - WIDTH*0.6)**2 + (y - HEIGHT*0.4)**2) / (100**2))
    else:
        # Winter: High pressure over interior
        pressure = 1020 - 10 * np.exp(-((x - WIDTH*0.6)**2 + (y - HEIGHT*0.4)**2) / (100**2))
    
    # Add noise for realism
    noise = np.random.normal(0, 0.5, (HEIGHT, WIDTH))
    pressure = np.clip(pressure + noise, 980, 1040)
    
    # Normalize to 0-255
    pressure_normalized = ((pressure - 980) / (1040 - 980) * 255).astype(np.uint8)
    img = Image.fromarray(pressure_normalized)
    img.save(f"sa_pressure_{month+1}.png")
    print(f" Saved South Africa pressure texture for month {month+1}")
    return img

def preview_weather(month):
    """Generate preview plots for all weather parameters"""
    plt.figure(figsize=(18, 12))
    current_month = datetime.now().month
    month = month if month else current_month - 1  # 0-based
    
    # Temperature Plot
    temp_img = generate_sa_temperature_texture(month)
    plt.subplot(2, 2, 1)
    plt.imshow(temp_img, cmap='coolwarm', extent=[16.45, 32.89, -34.83, -22.13])
    plt.colorbar(label='Temperature (째C)')
    plt.title(f'South Africa Temperature - Month {month+1}')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    
    # Precipitation Plot
    precip_img = generate_sa_precipitation_texture(month)
    plt.subplot(2, 2, 2)
    plt.imshow(precip_img, cmap='Blues', extent=[16.45, 32.89, -34.83, -22.13])
    plt.colorbar(label='Precipitation (mm)')
    plt.title(f'South Africa Precipitation - Month {month+1}')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    
    # Wind Plot
    wind_img = generate_sa_wind_texture(month)
    plt.subplot(2, 2, 3)
    plt.imshow(wind_img, extent=[16.45, 32.89, -34.83, -22.13])
    plt.title(f'South Africa Wind Patterns - Month {month+1}')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    
    # Pressure Plot
    pressure_img = generate_sa_pressure_texture(month)
    plt.subplot(2, 2, 4)
    plt.imshow(pressure_img, cmap='viridis', extent=[16.45, 32.89, -34.83, -22.13])
    plt.colorbar(label='Pressure (hPa)')
    plt.title(f'South Africa Pressure Systems - Month {month+1}')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    
    # Add city markers
    for station in SA_WEATHER_STATIONS:
        lon = SA_BOUNDS["min_lon"] + station["x"] * (SA_BOUNDS["max_lon"] - SA_BOUNDS["min_lon"])
        lat = SA_BOUNDS["min_lat"] + (1-station["y"]) * (SA_BOUNDS["max_lat"] - SA_BOUNDS["min_lat"])
        plt.scatter(lon, lat, c='black', s=30)
        plt.text(lon, lat, station["name"], fontsize=8)
    
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    print("Generating South Africa weather visualizations...")
    current_month = datetime.now().month - 1  # Convert to 0-based index
    preview_weather(current_month)
```

```python
# api/south_africa_pollution_api.py
from fastapi import FastAPI, Response
from fastapi.middleware.cors import CORSMiddleware
import numpy as np
from PIL import Image
import io
import uvicorn
from datetime import datetime

app = FastAPI(title="South Africa Pollution API")

# Enable CORS
app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],
    allow_methods=["GET"],
    allow_headers=["*"]
)

# South Africa bounding box
SA_BOUNDS = {
    "min_lon": 16.45,
    "max_lon": 32.89,
    "min_lat": -34.83,
    "max_lat": -22.13
}

# Major industrial areas in South Africa
INDUSTRIAL_SOURCES = [
    {"name": "Johannesburg", "coords": [28.0473, -26.2041], "intensity": 0.9},
    {"name": "Sasolburg", "coords": [27.8167, -26.8167], "intensity": 0.85},
    {"name": "Secunda", "coords": [29.1667, -26.5167], "intensity": 0.8},
    {"name": "Durban", "coords": [31.0292, -29.8587], "intensity": 0.75},
    {"name": "Cape Town", "coords": [18.4241, -33.9249], "intensity": 0.7}
]

def generate_sa_pm25_texture():
    width, height = 512, 512
    x, y = np.meshgrid(np.linspace(0, width-1, width), np.linspace(0, height-1, height))
    
    # Initialize empty pollution field
    pm25 = np.zeros((height, width))
    
    # Add pollution sources for each industrial area
    for source in INDUSTRIAL_SOURCES:
        # Normalize coordinates to texture space
        norm_x = (source["coords"][0] - SA_BOUNDS["min_lon"]) / (SA_BOUNDS["max_lon"] - SA_BOUNDS["min_lon"])
        norm_y = 1 - (source["coords"][1] - SA_BOUNDS["min_lat"]) / (SA_BOUNDS["max_lat"] - SA_BOUNDS["min_lat"])
        
        tex_x = norm_x * width
        tex_y = norm_y * height
        
        # Add gaussian plume for this source
        dist = np.sqrt((x - tex_x)**2 + (y - tex_y)**2)
        pm25 += source["intensity"] * np.exp(-dist / (50 + np.random.random() * 30))
    
    # Add seasonal variation (worse in winter)
    month = datetime.now().month
    season_factor = 0.7 + 0.3 * np.cos(2*np.pi*(month-6)/12)
    pm25 *= season_factor
    
    # Add some noise for realism
    noise = np.random.normal(0, 0.05, (height, width))
    pm25 = np.clip(pm25 + noise, 0, 1)
    
    # Convert to 8-bit
    pm25 = (pm25 * 255).astype(np.uint8)
    img = Image.fromarray(pm25)
    
    # Save to bytes
    img_bytes = io.BytesIO()
    img.save(img_bytes, format="PNG")
    img_bytes.seek(0)
    
    return img_bytes.getvalue()

def generate_sa_wind_texture():
    width, height = 512, 512
    x, y = np.meshgrid(np.linspace(0, width-1, width), np.linspace(0, height-1, height))
    
    # Prevailing winds in South Africa:
    # - Summer: SE winds
    # - Winter: NW winds
    month = datetime.now().month
    is_summer = month in [11, 0, 1, 2]  # Nov-Feb
    
    if is_summer:
        # Southeast winds
        u = np.ones((height, width)) * 0.7  # East component
        v = np.ones((height, width)) * -0.5  # South component
    else:
        # Northwest winds
        u = np.ones((height, width)) * -0.6  # West component
        v = np.ones((height, width)) * 0.4    # North component
    
    # Add some turbulence
    turbulence = np.sin(2*np.pi*x/width*4) * np.cos(2*np.pi*y/height*3) * 0.3
    u += turbulence
    v += turbulence * 0.7
    
    # Normalize to [0,255]
    r = ((u + 1) / 2 * 255).astype(np.uint8)  # East-West
    g = ((v + 1) / 2 * 255).astype(np.uint8)  # North-South
    b = np.zeros_like(r)                      # Unused
    
    wind_img = np.stack([r, g, b], axis=2)
    img = Image.fromarray(wind_img, 'RGB')
    
    # Save to bytes
    img_bytes = io.BytesIO()
    img.save(img_bytes, format="PNG")
    img_bytes.seek(0)
    
    return img_bytes.getvalue()

@app.get("/api/sa/pm25-texture")
async def get_sa_pm25_texture():
    """Returns PM2.5 data for South Africa as PNG"""
    return Response(content=generate_sa_pm25_texture(), media_type="image/png")

@app.get("/api/sa/wind-texture")
async def get_sa_wind_texture():
    """Returns wind data for South Africa as PNG"""
    return Response(content=generate_sa_wind_texture(), media_type="image/png")

@app.get("/api/sa/hotspots")
async def get_sa_hotspots():
    """Returns industrial hotspots in South Africa"""
    features = [
        {
            "type": "Feature",
            "properties": {
                "name": source["name"],
                "intensity": source["intensity"],
                "description": f"Industrial area in {source['name']}"
            },
            "geometry": {
                "type": "Point",
                "coordinates": source["coords"]
            }
        } for source in INDUSTRIAL_SOURCES
    ]
    
    return {
        "type": "FeatureCollection",
        "features": features
    }

if __name__ == "__main__":
    uvicorn.run(app, host="0.0.0.0", port=8000)
```
